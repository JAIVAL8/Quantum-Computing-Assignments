```python
from sympy import Matrix
import random

# counts dictionary from Qiskit result

counts = {'10110': 130, '11111': 131, '10010': 117, '00011': 128, '01010': 127, '01001': 128, '00100': 117, '11100': 115, '10101': 138, '00000': 107, '11000': 128, '10001': 130, '00111': 129, '01101': 142, '01110': 118, '11011': 115}
n = 5

# remove trivial result
bitstrings = [k for k in counts.keys() if k != "0"*n]

# shuffle to avoid dependent rows
random.shuffle(bitstrings)

# take first n-1 equations
selected = bitstrings[:n-1]

print("Selected measurement strings:")
print(selected)

# build matrix U
U = Matrix([[int(b) for b in s] for s in selected])

print("\nMatrix U:")
print(U)

# compute nullspace
ns = U.nullspace()

print("\nRecovered secret string:")

for vec in ns:
    s = [int(v % 2) for v in vec]
    print("".join(map(str,s)))
```
